```python
x=df["date"]
y=df["Altura(mm)"]

plt.title("Mareas, Isla de Cedros. Marzo 2016 ")
plt.ylabel('Altura del mar (mm)')
plt.xlabel('Fecha')
plt.grid(True)


plt.plot(x,y)
plt.show()
```
